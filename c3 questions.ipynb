# 1.6 loc vs iloc Methods

* we have seen that when we appended a new dataframe into an existing one that the row index numbers were sequencially starting from 0 up to the length of the dataframe when the `ignore_index` was `True`. however, the row index numbers started from 0 again at every original dataframe before appending them when the `ignore_index` was set to `False`. what are the implications of this you? well, extracting a row using `iloc` method for row index 0 for example, you see in the output of the code that both result in the same extracted row although the dataframe on the right has two rows with indices 0. Then why both zero indexed rows at the right dataframes where not extracted? this is because iloc method always gets regards row index numbers starting from zero up to the length of the dataframe regardless of the row index label. for example, when row index label is 0 at the row where name is bob in the right dataframe, the row index number is still considered to 4 by iloc method although its row index label is 0. now have do you reset the row index labels to match the row index numbers? well, you need to apply the reset index method on the dataframe. 

Screen Shot 2022-09-21 at 8.08.01 PM.png

* This is the syntax of the reset index method applied on a dataframe. There are two options when it comes to using this syntax. you do not need to apply reset index method on a dataframe when using the append method as you can already set the `ignore_index` to `True`. 

  ```python
  new_df = new_df.reset_index()
  ```
  or
  ```python
  new_df.reset_index(inplace=True)
  ```

````python
dataset = {'name': ['John', 'Tom', 'Sara', 'Evan'],
           'position': ['engineer', 'artist', 'architect', 'recruiter'],
           'age': [28, 29, 27.0, 31],
           'experience': [2.5, 5, 7, 10]}

df = pd.DataFrame(dataset) # convert dataset dictionary to dataframe

new_rows = {'name': ['Bob', 'Mary'],
           'position': ['technician', 'data scientist'],
           'age': [35, 25],
           'experience': [15, 1.5]}

new_rows = pd.DataFrame(new_rows) # convert new_rows dictionary to dataframe
````

````python
new_df = df.append(new_rows, ignore_index=False)
print(new_df.iloc[0],'\n') 
print(new_df.iloc[4])

````

* now what if you want to really search by the row index label not the row index number? here is where loc method is beneficial. the iloc method will extract a row based on the row index number, but the loc method will extract a row based on the row index label. for example, if you use the iloc method to extract row at row index number 0, you will get the first row. however, if you apply the loc method on the dataframe with row index label 0, you will extract two rows since two rows exist with the same row index labels. you can also extract multiple row index labels as such. furthermore, you can slice a dataframe with the loc method, but you need to index sort the dataframe before doing so. 

Screen Shot 2022-09-21 at 8.26.56 PM.png

````python
dataset = {'name': ['John', 'Tom', 'Sara', 'Evan'],
           'position': ['engineer', 'artist', 'architect', 'recruiter'],
           'age': [28, 29, 27.0, 31],
           'experience': [2.5, 5, 7, 10]}

df = pd.DataFrame(dataset) # convert dataset dictionary to dataframe

new_rows = {'name': ['Bob', 'Mary'],
           'position': ['technician', 'data scientist'],
           'age': [35, 25],
           'experience': [15, 1.5]}

new_rows = pd.DataFrame(new_rows) # convert new_rows dictionary to dataframe

new_df = df.append(new_rows, ignore_index=False)

print(new_df.loc[0],'\n') # using loc method 

print(new_df.iloc[0]) # using iloc method
````

# 1.8 Sorting DataFrames

* sorting dataframes can be also done using pandas. using the `sort_values()` method, you can re-arrange the values of a dataframe by sorting a specfic column in a dataframe. Here is the syntax of the `sort_values()` method:

  ```python
  DataFrame.sort_values(by, ascending=True, inplace=False, ignore_index=False, na_position='last')
  ```

* Here is the breakdown of the method's arguments:

  * `by`: name or a list of names to sort by.
  * `ascending`: this argument will sort ascending if `True`, descending otherwise. `True` is defualt value.
  * `inplace`: if True, then it performs operation with updating values of the same dataframe variable applied on. `False` is defualt value.
  * `ignore_index` if True, the resulting dataframe will be labeled as such 0, 1, 2, ... , n-1. `False` is defualt value.
  * `na_position`: puts NaN values at the begining of dataframe if `first`, and `last` will keep it at the end of the dataframe. 

* note: NaN value represents a missing value in dataset.

* below, `df` dataframe is defined with three columns of letters and digits. also, there is a missing value represented by NaN. notice `col1` has three 

````python
df = pd.DataFrame({
    'col1': ['b', 'a', 'a', np.nan, 'a', 'c'],
    'col2': [4, 0, 2, 9, 1, 3],
    'col3': [2, 5, 1, 8, 7, 10]
})

df
````

* using the `sort_values()` method, the dataframe can be sorted using a single column (`col1` for example) while keeping NaN value at the begining of the dataframe.

````python
df.sort_values(by=['col1'], na_position='first', inplace=False)
````

* now, how does the `sort_values()` sort by two columns for example `col1` and `col2`? notice `col1` has three rows with the same value `a`. In this case, `sort_values()` will sort column `col1` first then `col2` without messing up with the sorted order of `col1`. notice that `col2` values were sorted only where `col1` has values `a`. 

````python
df.sort_values(by=['col1', 'col2'], na_position='first', inplace=False)
````

# 1.9 Replacing Values




* with the help of `DataFrame.replace()` method, the values in a dataframe can be replaced with different values. here is the syntax of this function:

  ```python
  DataFrame.replace(to_repalce, value, inplace=False)
  ```
* here is the breakdown of the input arguments:
  * `to_replace`: specifies how to find the values to be replaced. values could be scalar string or numeric values, a list of string and/or numeric values or a dictionary of string and/or numeric values.
  * `value`: is the replacing values that matches with `to_replace`. it could also be a scalar, list or a dictionary of replacing values. 
  * `inplace`: whether to modify the existing dataframe instead of creating new one. defualt value is `False`.

* to illustrate the replace method, the following dataframe will be used:    

````python
df = pd.DataFrame({
    'col1': ['b', 'a', 'a', np.nan, 'a', 'c'],
    'col2': [4, 0, 2, 9, 1, 3],
    'col3': [2, 5, 1, 0, 7, 10],
    'col4': [7, 5, 3, 5, 3, np.nan]
})

df
````

* In this example, the NaN values (represented using NaN numpy value np.nan) is replaced with the string value `missing value` while keeping the `inplace` flag to `False` to avoid updating the `df` dataframe itself.

````python
x = df.replace(np.nan, 'missing value')
x
````

* similarly, you can replace multiple values in list-like `to_replace` into a singular value. notice that the changes happened only columns `col2`, `col3` and `col4`.

````python
x = df.replace([0, 1, 5], 'X')
x
````

# 1.13 Joining DataFrame

* The `DataFrame.join()` method in pandas allows to join columns of another dataframe to an existing one. it is possible to join columns with other dataframe either on index or on a key column. Here is the syntax of the join method:

  ```python
  DataFrame.join(other, on=None, how='left', lsuffix='', rsuffix='')
  ```

* This is the breakdown of the method:
  * `other`: can be a DataFrame, Series or a list containing any combination of them to join with an existing dataframe. 
  * `on`: column or index level names in the existing dataframe to join on the in index in `other`. This is similar to VLOOKUP in Excel. this is. anoptional argument.
  * `how`: is a string value that specifies how to handle the operation of joining two dataframes:
    * `left`: using the the existing frame's index or column when joining. this is the default value.
    * `right`: using `other` frame's index or column when joining.
    * `outer`: form a unoin of the existing frame and the `other` and sort it.
    * `inner`: form an intersection of the existing frame with `other` while preserving the order of the existing frame.
  * `lsuffix`: suffix to use from left frame’s overlapping columns. 
  * `rsuffix`: suffix to use from right frame’s overlapping columns.

# 1.16 Pivot Tables in Pandas

* the `DataFrame.pivot()` creates a spreadsheet-like pivot table as a dataframe. this is used to reshape a given dataframe organized by a specified index and column values. the resulting dataframe will have a `MultiIndex` in the column. here is the syntax of the method:

  ```python
  DataFrame.pivot(index=None, columns=None, values=None, dropna=True)
  ```
* here is the breakdown of the method:
  * `index`: is a column to be used as an index in the pivot table. 
  * `columns`: is a column whose values are used as the columns names in the resulting pivot table. 
  * `values`: column(s) to be used to fill the dataframe's values corresponding to `index` and `columns`.  If not specified, all remaining columns will be used as values. 
  * `dropna`: do not include columns whose entries are all NaN values. 

* as can be seen in the figure below, `col1` has two unique values which are used as row index values in the resulting pivot table. `col2` values are used as column names in the resulting pivot table with a total of three unique values. `col3` and `col4` are used to fill the values of the pivot tables corresponding row index values and column names in the resulting pivot table.  

Screen Shot 2022-12-04 at 6.41.43 PM.png

````python
df = pd.DataFrame({'col1': ['one', 'one', 'one', 'two', 'two',
                           'two'],
                   'col2': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'col3': [0, 1, 2, 3, 4, 5],
                   'col4': ['a', 'b', 'c', 'd', 'e', 'f']})
df
````

* notice that the resulting pivot table has column names `A`, `B` and `C`. However, those column names are different since the first set of `A`, `B` and `C` columns are sourced from `col3` are noted above the columns names. also, the second set of `A`, `B` and `C` columns names are sourced from `col4`. 

````python
df_pivot = df.pivot(index='col1', columns='col2', values=['col3','col4'])
df_pivot
````

* now the question is: how to access column values in the resulting pivot table as each column is denoted by column name and source column? The answer to this is that a tuple with column source and column name must be used to access column values in the resulting pivot table. 

  ```python
  ('source_col','col_name')
  ```

* in this example, column `A` sourced from `col3` is accessed in the pivot table.

````python
df_pivot[('col3','A')]
````

### Creating Data and Time Data

* a data and time data can be created using the `Timestamp()` function where the year, month, day, hour, minute and second can be assigned. Here are different techniques to do it.

* similarly, a date range can be created using the `pd.date_range()` method in pandas. here is the syntax of the method:

  ```python 
  pd.date_range(start=None, end=None, periods=None, freq='D')
  ```
* here is the breakdown of the method:
  * `start`: Left bound for generating dates.
  * `end`: Right bound for generating dates.

  * `freq`: a string denoting the frequency of generated dates:
    * `D`: days frequency (default value)

    * `H`: hourly frequncy.
    * `S`: secondly frequency.
    * chech documentation for more.

### Boolean Operations with DateTime

* DateTime indexed dataframes can also be sliced using boolean expressions with the use of `loc()` method.

````python
dt = pd.date_range(start='2022-01-01', end='2022-01-5', freq='6H')
df = pd.DataFrame({'col1':dt,
                   'col2':list(range(len(dt)))})
df.set_index('col1', inplace=True)
df
````

* in the example below:
  * the condition is mask for the `loc()` method to extract the rows where `condition` returns `True`.
  * The `pd.Timestamp()` method is used to create an instance of a single DateTime for boolean condition expression. 
  * the `df.index` extracts all dates in the row index labels for comparison with the singluar DateTIme generated by `pd.Timestamp()` method. 
  * the example extracts all rows where DateTime is greater than `2022-01-02 12:00:00`.

````python
condition = (df.index > pd.Timestamp('2022-01-02 12:00:00'))
df.loc[condition]
````

* the example below extracts rows where row index labels are less than `2022-01-03 12:00:00` and greater than `2022-01-01 18:00:00`.


````python
condition = (df.index < pd.Timestamp('2022-01-03 12:00:00')) & (df.index > pd.Timestamp('2022-01-01 18:00:00'))
df.loc[condition]
````
